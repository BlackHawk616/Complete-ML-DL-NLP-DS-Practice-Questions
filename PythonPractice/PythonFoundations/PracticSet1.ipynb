```python
# Number Guessing Game
import random

def main():
    print("Welcome To Number Guessing Game!")
    print("Guess a number between 1 and 10")
    number = random.randint(1,10)
    choice = int(input("Enter Your Guess : "))

    if choice == number:
        print("Congratulations! You guessed the number correctly.")
    else:
        print(f"Sorry! The correct Number Was {number}")

if __name__ == "__main__":
    main()
    
```

```python
# Simple Calculator

def calculator():
    print("Welcome To Simple Calculator!")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    choice = int(input("Enter Your Choice : "))
    print("Enter Two Numbers : ")
    num1 = float(input("First Number : "))
    num2 = float(input("Second Number : "))
    if choice == 1:
        print(f"The Sum of {num1} and {num2} is {num1 + num2}")
    elif choice == 2:
        print(f"The Difference of {num1} and {num2} is {num1 - num2}")
    elif choice == 3:
        print(f"The Product of {num1} and {num2} is {num1 * num2}")
    elif choice == 4:
        if num2 != 0:
            print(f"The Quotient of {num1} and {num2} is {num1 / num2}")
        else:
            print("Error: Division by zero is not allowed.")
    else:
        print("Invalid Choice! Please select a valid operation.")
```

```python
# Palindrome Checker

def palindrom_checker():
    print("Welcome to palindrome checker!")
    word = input("Enter a word : ")
    if word == word[::-1]:
        print(f"{word} is a palindrome")
    else:
        print(f"{word} is not a palindrome")
```

```python
# Primer Number Checker

def prime_number_checker():
    print("Welcome to prime Number checker ")
    num = int(input("Enter a number : "))
    if num > 1:
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                print(f"{num} is not a prime number")
                break
        else:
            print(f"{num} is a prime number")

prime_number_checker()
```

```python
# Fibonacci Sequence geneartor

def fibonacci_sequence_generator():
    print("Welcome to fibonacci sequence generator!")
    n = int(input("Enter the number of terms : "))
    a = 0
    b = 1
    for i in range(n):
        print(a, end='')
        a,b = b, a+b
fibonacci_sequence_generator()

# Using Recursive Function to print Fibonacci Sequence
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
# Word Frequency Counter

def word_frequency_counter():
    print("Welcome to word frequency counter!")
    text = input("Enter a string : ")
    words = text.split()
    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    print("Word frequencies:")
    for word, count in frequency.items():
        print(f"{word}: {count}")

word_frequency_counter()
```
